## Data Cleaning
### Objective

Prepare the dataset for analysis by handling missing values and creating meaningful features.

### Tasks Performed

- Removed irrelevant columns (Agent, Company,Personal Info)
- Handled missing values in Country and Children
- Created Arrival Date column
- Renamed columns for readability
- Checked duplicates
- Verified data quality

```python
df.isnull().sum()
```

```python
df["children"] = df["children"].fillna(0)
```

```python
df["country"] = df["country"].fillna("Unknown")
```

```python
df = df.drop(columns=["agent","company"])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns = ["name","email","phone-number","credit_card"])
```

```python
df["children"] = df["children"].astype(int)
```

```python
df["Arrived Dates"] = pd.to_datetime(df["arrival_date_year"].astype(str) + " " + df["arrival_date_month"].astype(str) + " " + df["arrival_date_day_of_month"].astype(str))
```

```python
df["Arrived Dates"] = df["Arrived Dates"].dt.strftime("%d-%m-%Y")
```

```python
df = df.drop(columns = ["arrival_date_week_number","arrival_date_day_of_month"])
```

## Feature Engineering
### Objective

Create new variables to improve business analysis.

### Features Created

- Total Stay Nights:
Calculated total stay duration using weekend and weekday nights.

- Lead Time Category :
Classified bookings into:
 Early Booking
,Advance Booking
and Last Minute Booking

- Room Change Status:
Identified whether assigned room differed from reserved room

```python
df.rename(columns = {"is_canceled":"booking_status"},inplace = True)
```

```python
df["booking_status"] = df["booking_status"].map({0:"Not Cancelled",1:"Cancelled"})
```

```python
df["Total_stays"] = df["stays_in_week_nights"] + df["stays_in_weekend_nights"]
```

```python
df["Category"] = np.where(df["lead_time"] >= 91,"Early Booking",np.where((df["lead_time"] <= 90) & (df["lead_time"]>=8),"Advance Booking","Last Minute Booking"))
```

```python
df["room_change_flag"] = np.where(df["reserved_room_type"] == df["assigned_room_type"],"Not Changed","Changed")
```

```python
df["room_change_flag"] = np.where(df["reserved_room_type"] == df["assigned_room_type"],"Not Changed","Changed")
```

## Exploratory Data Analysis (EDA)

A) What are the booking patterns of customers ?

```python
hotel_count = df["hotel"].value_counts()
plt.title("Booking Distribution by Hotel Type")
plt.pie(hotel_count,labels = hotel_count.index, autopct="%1.1f%%")
```

```python
Booking_type = df["Category"].value_counts() 
plt.title("Booking Distribution by Booking Channel")
plt.bar(Booking_type.index,Booking_type.values)
plt.xlabel("Booking Type")
plt.ylabel("No of Booking")
```

```python
Market_Category = df["market_segment"].value_counts()
plt.title("Bookings by Market Segment")
plt.bar(Market_Category.index,Market_Category.values) 
plt.xlabel("Market_Category") 
plt.ylabel("No of Booking")
plt.xticks(rotation = 20)
```

```python
Top10_Country_Booking = df["country"].value_counts().head(10) 
plt.title("Top Countries by Number of Bookings")
plt.bar(Top10_Country_Booking.index,Top10_Country_Booking.values)
plt.xlabel("Countrys") 
plt.ylabel("No of Booking")
```

```python
Table = pd.pivot_table(df,index="arrival_date_month",columns="arrival_date_year",aggfunc="size")
month_order = ["January","February","March","April","May","June","July","August","September","October","November","December"]
Table = Table.reindex(month_order) 

plt.figure(figsize = (10,5))
for year in Table.columns:
    plt.plot(Table.index,Table[year],marker = "o",label = year) 
plt.title("Monthly Booking Trend by Year")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

### Key Insights / Observations
- City Hotels account for around 66% of total bookings, making them the primary revenue source.
- Most customers book well in advance (Early & Advance bookings) rather than at the last minute.
- Online Travel Agencies (OTAs) generate the largest share of bookings, showing customers prefer online platforms.
- Portugal contributes the highest number of bookings, followed by the UK, France, Spain, and Germany.
- Hotel demand is seasonal, with bookings increasing significantly during holiday and summer months

B) Which factors are associated with higher cancellation rates ?

```python
plt.subplot(2,2,1)
plt.title("City Hotel")
Table = pd.crosstab(df["hotel"],df["booking_status"]) 
plt.pie(Table.loc["City Hotel"],labels = Table.columns,autopct="%1.1f%%")

plt.subplot(2,2,2)
plt.title("Resort Hotel") 
plt.pie(Table.loc["Resort Hotel"],labels = Table.columns,autopct = "%1.1f%%")
```

```python
Table = pd.crosstab(df["booking_status"],df["Category"])
Table 
```

```python
Table = pd.crosstab(df["booking_status"],df["Category"])
pos = np.arange(len(Table.index)) 
Bandwidth = 0.25 
br = pos 
br1 = br + Bandwidth 
br2 = br1 + Bandwidth
plt.bar(br,Table["Advance Booking"],width=Bandwidth,label="Advance Booking") 
plt.bar(br1,Table["Early Booking"],width=Bandwidth,label="Early Booking")
plt.bar(br2,Table["Last Minute Booking"],width = Bandwidth,label = "Last Minute Booking")
plt.legend() 
plt.title("Booking Category Distribution by Booking Status") 
plt.xticks(br+Bandwidth,Table.index) 
plt.xlabel("Booking Status") 
plt.ylabel("No of Booking")
```

```python
Table = pd.crosstab(df["market_segment"],df["booking_status"])
pos = np.arange(len(Table.index))
Bandwidth = 0.25 
br = pos 
br1 = pos + Bandwidth 
plt.bar(br,Table["Cancelled"],width = Bandwidth,label="Cancelled")
plt.bar(br1,Table["Not Cancelled"],width = Bandwidth,label = "Not Cancelled")
plt.xticks(br+Bandwidth,Table.index) 
plt.xticks(rotation = 20) 
plt.xlabel("Market Segment")
plt.ylabel("Booking Count")
plt.title("Cancellation Analysis by Market Segment") 
plt.legend()
```

```python
Table = pd.crosstab(df["booking_status"],df["room_change_flag"])
pos = np.arange(len(Table.index)) 
Bandwidth = 0.25 
br = pos 
br1 = br + Bandwidth 
plt.bar(br,Table["Changed"],width = Bandwidth,label = "Changed")
plt.bar(br1,Table["Not Changed"],width = Bandwidth,label = "Not Changed")
plt.legend()
plt.title("Room Change Analysis by Booking Status")
plt.xlabel("Booking Status")
plt.xticks(br+Bandwidth,Table.index) 
plt.ylabel("No of Booking")
```

## Key Insights / Observations
- City Hotels have a much higher cancellation rate than Resort Hotels.
- Bookings made far in advance (high lead time) are more likely to be canceled.
- Customers booking through Online Travel Agencies show relatively higher cancellation rates.
- Guests with repeated cancellations in the past are more likely to cancel future bookings.
- Non-refundable bookings have significantly lower cancellation rates.

## Business Recommendation
- Introduce flexible deposit policies for high-risk bookings.
- Use predictive analytics to identify likely cancellations.
- Encourage non-refundable or partially refundable booking options.

What factors influence the Average Daily Rate (ADR) of hotel bookings?

```python
Table = round(df.groupby("hotel")["adr"].mean(),2)
plt.bar(Table.index,Table.values)
plt.xlabel("Hotel Type")
plt.ylabel("Average Daily Rate (ADR)")
plt.title("Average Daily Rate by Hotel Type")
```

```python
Table = round(df.groupby('market_segment')['adr'].mean(),2)
plt.bar(Table.index,Table.values)
plt.xticks(rotation = 20)
plt.title("Average Daily Rate by Market Segment")
plt.xlabel("Market Segment")
plt.ylabel("Average Daily Rate (ADR)")
```

```python
Table = pd.pivot_table(df,values="adr",index="arrival_date_month",columns="arrival_date_year",aggfunc="mean") 
month_order = ["January","February","March","April","May","June","July","August","September","October","November","December"]
Table = Table.reindex(month_order)
plt.figure(figsize=(10,5))

for year in Table.columns:
    plt.plot(Table.index,Table[year],marker="o",label=year)

plt.title("Average Daily Rate Trend by Month and Year")
plt.xlabel("Month")
plt.ylabel("Average Daily Rate (ADR)")
plt.xticks(rotation=45)
plt.legend(title="Year")
plt.tight_layout()
plt.show()
```

## Key Insights / Observations
- ADR increases during peak travel seasons because of higher customer demand.
- City Hotels generally charge a higher ADR due to greater business and urban travel demand.
- ADR varies across market segments, with corporate and direct bookings often generating higher room rates.
- Special requests and premium room preferences are associated with higher ADR.
